```python
#Importamos librerias(pip install tensorflow)
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model

# Cargamos el modelo HDF5
model = load_model("fire_detection_model.h5")
```

```python
# Procesa la imagen
# Cambiar el origen de la imagen
imagen = load_img(r'C:\Users\nicom\Untitled Folder\fireDetencion\test\no_fire\NF_14.jpg', target_size=(224, 224))
imagen_array = img_to_array(imagen) / 255.0  # Ponemos los valores entre 0 y 1
imagen_array = imagen_array.reshape((1, 224, 224, 3)) # Resizeamos la imagen
resultado = model.predict(imagen_array)
if resultado[0] > 0.5:
    print("Incendio detectado")
else:
    print("No hay incendio")
```

```python
# Por si queremos ver la imagen
import matplotlib.pyplot as plt
plt.imshow(imagen)
plt.title("Predicción: Incendio" if resultado[0] > 0.5 else "Predicción: No hay incendio")
plt.axis('off')
plt.show()
```
